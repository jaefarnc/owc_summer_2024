```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
base="ivory"
```

```python
#df=pd.read_csv(f"./{base}/{base}_feature_engineered.csv")
df = pd.read_csv(f'{base}/{base}_usercount_classifier0.csv')
```

```python
df['proportion_flagged'] = df['num_flagged_tweets']/df['num_scraped_tweets']
columns = ['handle', 'followers_count','post_freq','std_dev','avg_tweet_length', 'proportion_flagged']
df = df[columns]
```

```python
df.head()
```

```python
df.dtypes
```

```python
columns = ['followers_count','post_freq','std_dev','avg_tweet_length']
column = columns[0]
def floatify(row):
    temp = str(row[column]).replace(',','')
    return float(temp)
df[column]=df.apply(floatify,axis=1)

df.head()
```

```python
## we can plot each distribution to fit it to the actual values
import pandas as pd
from sklearn.preprocessing import RobustScaler, StandardScaler, QuantileTransformer
columns = ['followers_count','post_freq','std_dev','avg_tweet_length','proportion_flagged']
for column in columns:
    qt = QuantileTransformer(output_distribution='normal')
    df[[column]]=qt.fit_transform(df[[column]])
    # scaler = StandardScaler()
    # df[column]=scaler.fit_transform(df[[column]])  # Fit only on the column you want to normalize
```

```python
for column in columns:
    print(column)
    print(df[column].min(),df[column].max())
```

```python
df.head()
```

```python
##Convert to absolute values for the desired ones
#followers_count: low
df['followers_count'] = -df['followers_count']
```

```python
#post_freq: extremes
columns = ['post_freq','std_dev','avg_tweet_length']
for column in columns:
    df[column] = abs(df[column])
```

```python
columns = ['followers_count','post_freq','std_dev','avg_tweet_length','proportion_flagged']
for column in columns:
    qt = QuantileTransformer(output_distribution='normal')
    df[[column]]=qt.fit_transform(df[[column]])
    # scaler = StandardScaler()
    # df[column]=scaler.fit_transform(df[[column]])  # Fit only on the column you want to normalize
```

```python
df.head()
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=1)
columns = ['followers_count','post_freq','std_dev','avg_tweet_length', 'proportion_flagged']

df['pca_score'] = pca.fit_transform(df[columns])
df['pca_rank'] = df['pca_score'].rank(ascending=False)
```

```python
df.head()
```

```python
sorted_df = df.sort_values(by='pca_rank', ascending=True)
```
